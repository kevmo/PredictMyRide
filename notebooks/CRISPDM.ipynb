```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/msrp.csv')
```

```python
df.head()
```

## Preprocessing

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df.columns
```

```python
df.head()
```

```python
df.dtypes
```

```python
col_strings = df.dtypes[df.dtypes == 'object'].index
col_strings
```

```python
for col in col_strings:
    df[col]= df[col].str.lower().str.replace(' ', '_')
```

```python
df.head()
```

## Exploratory Data Analysis

```python
for col in df.columns:
    print("#### COLUMN: ", col)
    print("UNIQUE: ", df[col].nunique())
    print(df[col].unique()[:5])
    print("#####")
```
